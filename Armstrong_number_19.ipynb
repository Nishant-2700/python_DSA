```python
#Checking Armstrong Number

num=int(input("Enter a number"))
temp = num
digit = len(str(num))
ans=0

while (temp>0):
    temp = temp%10
    ans = temp ** digit
    tem = temp//10

    
if ans==temp:
    print("number is Armstrong")
else:
    print("not")
```

```python
#Print the multiple of 10, in a given range entered by the user. (Note: Do this question using range())

start =int(input("Enter a starting number"))
end = int(input("Enter a ending number"))
for i in range(start,end+1):
    if i%10==0:
        print(i)
```
