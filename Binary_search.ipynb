```python
def binarysearch(arr,ele,si,ei):
    if (si>ei):
        return -1
    mid = (si+ei)//2
    if(ele ==arr[mid]):
        return mid
    elif (ele < arr[mid]):
        return binarysearch(arr,ele,si,mid-1)
    else:
        return binarysearch(arr,ele,mid+1,ei)
ans = binarysearch([1,2,3,4,5,6,7,8,9],8,0,10)
print(ans)
```
