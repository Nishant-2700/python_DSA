```python
#reverse of list
l = [1,2,3,4,5,6]
l1 = len(l)
for i in range(l1//2):
    l[i], l[l1 - i - 1] = l[l1 - i - 1], l[i]
print(l)
```

```python
#Star Square

n = int(input("Enter a number"))
i=0
while i<n:
    j=0
    while j<n:
        print("*", end = " ")
        j+=1
    i+=1
    print()
```

```python
#Star Right Triangle

n = int(input("Enter a number"))
i=0
while i<n:
    j=0
    while j<=i:
        print("*", end = " ")
        j+=1
    i+=1
    print()
```

```python
#Number Pattern (Constant)

n = int(input("Enter a number"))
i=0
while i<n:
    j=0
    while j<=i:
        print("1", end = " ")
        j+=1
    i+=1
    print()
```

```python
#Increasing Number Pattern

n = int(input("Enter a number"))
i=1
while i<=n:
    j=1
    while j<=i:
        print(j, end = " ")
        j+=1
    i+=1
    print()
```

```python
#Alphabet patterns

n = int(input("Enter a number"))
i=1
while i<=n:
    j=0
    while j<i:
        print(chr(65+j), end = " ")
        j+=1
    i+=1
    print()
```

```python
# take input in a list

n = int(input("Enter a number of element"))
list = []
for i in range(n):
    x=int(input("Enter element"))
    list.append(x)
print(list)
```
