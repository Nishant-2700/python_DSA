```python
#Find occurrence of Char in a string recursively
def count(s, ch):
    if len(s) == 0:
        return 0
    if s[0] == ch:
        return 1 + count(s[1:], ch)
    else:
        return count(s[1:], ch)


string = input("Enter string: ")
char = input("Enter character to count: ")

result = count(string, char)
print("Total :", result)
```

```python
#return the index of char in a string recursively
def first(s, ch, i=0):
    if i == len(s):
        return -1
    if s[i] == ch:
        return i
    return first(s, ch, i + 1)


string = input("Enter string: ")
char = input("Enter character to find index: ")

result = first(string, char)
print("First index:", result)
```

```python
#Remove duplicate in a string
def remove(s):
    if len(s) <= 1:
        return s
    
    small_output = remove(s[1:])
    
    if s[0] in small_output:
        return small_output
    else:
        return s[0] + small_output


string = input("Enter string: ")

result = remove(string)
print("String after removing duplicates:", result)
```

```python
#replace the char question in string
def replace(s, old, new):
    if len(s) == 0:
        return ""
    if s[0] == old:
        return new + replace(s[1:], old, new)
    else:
        return s[0] + replace(s[1:], old, new)


string = input("Enter string: ")
old_char = input("Enter character to replace: ")
new_char = input("Enter new character: ")

result = replace(string, old_char, new_char)
print("Updated string:", result)
```
