```python
#Selection Sort

def selectionSort(arr):
    for i in range(len(arr)):
        min_index = i
        for j in range(i+1, len(arr)):
            if arr[min_index] > arr[j]:
                min_index = j
        #SWAP - min_index and i
        arr[min_index], arr[i] = arr[i], arr[min_index]
arr =[37, 2,4, 25,5,900, 36,8,11,12,17, 80,29,56,60,91]
selectionSort(arr)
print(arr) 
      
```
