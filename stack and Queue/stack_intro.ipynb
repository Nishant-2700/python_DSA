```python
#INBUILT QUEUE
import queue
q = queue.Queue()
q.put(1)
q.put(9)
q.put(0)

while not q.empty():
    print(q.get())
```

```python
#INBUILT STACK

import queue
s = queue.LifoQueue()
s.put(8)
s.put(5)
s.put(56)
s.put(59)

while not s.empty():
    print(s.get())
```
