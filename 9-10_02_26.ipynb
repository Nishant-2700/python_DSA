```python
#Replace Pi with 3.14
def replace_pi(s):
    if len(s) <= 0:# and len(s) == 1:
        return s
    if s[0] == "p" and s[1] =="i":
        return "3.14" + replace_pi(s[2:])
    else:
        return s[0] + replace_pi(s[1:])

replace_pi('abpicpi')
```

```python
#tower og hanoi
def tower_of_hanoi(n, source, aux, dest):
    # a = source, b = aux, c = dest
    #base case
    if n == 1:
        print('move 1st disk from', source, 'to', dest)
        return
    tower_of_hanoi(n-1, source, dest, aux)
    print('move', n, "th disk from", source, 'to', dest)
    tower_of_hanoi(n-1, aux, source, dest)

tower_of_hanoi(4, 'source', 'aux', 'dest')
```

```python
#Binary Search Recursive

def binaryRecursive(arr, k, si, ei):
    if si > ei:
        return -1

    mid = (si + ei) // 2
    if arr[mid] == k:
        return mid
    elif arr[mid] < k:
        return binaryRecursive(arr, k, mid + 1, ei)
    else:
        return binaryRecursive(arr, k, si, mid - 1)

ans = binaryRecursive([1,2,4,6,7,8,12,15,16,20,23], 15, 0, 10)
print(ans)
```

```python
def merge(a1, a2, a):
    i = 0
    j = 0
    k = 0
    while(i < len(a1) and j < len(a2)):
        if(a1[i] < a2[j]):
            a[k] = a1[i]
            k += 1
            i += 1
        else:
            a[k] = a2[j]
            k += 1
            j += 1
    while i < len(a1):
        a[k] = a1[i]
        k += 1
        i += 1
    while j < len(a2):
        a[k] = a2[j]
        k += 1
        j += 1
```

```python
def mergeSort(a):
    
    if(len(a) == 0 or len(a) == 1):
        return
    
    mid = len(a)//2
    a1 = a[0:mid]
    a2 = a[mid:]
    
    mergeSort(a1)
    mergeSort(a2)
    
    merge(a1, a2, a)
```

```python
arr = [5, 2, 9, 1, 6, 3]
mergeSort(arr)
print(arr)

```
