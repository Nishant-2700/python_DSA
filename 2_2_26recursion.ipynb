```python
#program to printN to 1


def printnto1(n):
    if n==0:
        return     
    print(n)
    
    printnto1(n-1)
    return
    
n = int(input())
printnto1(n)
```

```python
#progrma to print the sum of first N natural number

def sum(n):
    if n == 1:
        return 1
    smalloutput = sum(n-1)
    return smalloutput + n
    
n = int(input())
sum(n)
```

```python
#WAP to find the sum of all element of the array/list 

def sumoflist(lst):
    if len(lst) == 1:
        return lst[0]
    smallOutput = sumoflist(lst[1:])
    return smallOutput + lst[0]




lst = [2,5,3,6,3,6]
sumoflist(lst)
```
